```python
all_countries = list(covid_dataset_aggregated.index)

max_infection_rates = []

for c in all_countries:
    max_infection_rates.append(covid_dataset_aggregated.loc[c].diff().max())
covid_dataset_aggregated["Max_infection_rates"] = max_infection_rates
    
```

```python
covid_dataset_aggregated
```

```python
covid19_data = pd.DataFrame(covid_dataset_aggregated["Max_infection_rates"])
```
